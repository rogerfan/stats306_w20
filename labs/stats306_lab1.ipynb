```r
ggplot(data=diamonds, mapping=aes(x=carat, y=price)) + 
    geom_point()
```

```r
ggplot(data=diamonds, aes(x=carat, y=price, color=clarity)) + 
    geom_point(alpha=0.6) +
    facet_wrap(~cut, ncol=3) +
    labs(x='Carats', y='Price($)') +
    ggtitle('Diamond size by carat')
```
